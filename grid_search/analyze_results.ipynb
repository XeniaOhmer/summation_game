```python
import pandas as pd
```

```python
{
  "N": [20, 60],
  "loss_type": ["mse", "cross-entropy"],
  "n_epoch": [100],
  "batch_size": [32],
  "lr": [0.001],
  "receiver_embed_dim": [32, 64, 128],
  "temperature": [1, 1.25, 1.5],
  "temp_decay": [1.0, 0.995],
  "one_hot": [0, 1],
  "n_layers": [2, 3],
  "wrap_receiver": [1]
}
```

```python
results = {"N": [], 
           "loss_type": [], 
           "embedding_dim": [], 
           "temperature": [],
           "temp_decay": [],
           "n_layers": [],
           "one_hot": [],
           "one_hot": [],
           "loss_train": [], 
           "loss_val": [], 
           "aux_train": [], 
           "aux_val": []}

run_count = 0

for N in [20]: 
    for loss in ["mse", "cross-entropy"]:
        for receiver_embed_dim in [32, 64, 128]: 
            for temperature in [1, 1.25, 1.5]:
                for temp_decay in [1., 0.995]:
                    for one_hot in [0, 1]:
                        for n_layers in [2, 3]:
                            
                            results["N"].append(N)
                            results["loss_type"].append(loss)
                            results["embedding_dim"].append(receiver_embed_dim)
                            results["temperature"].append(temperature)
                            results["temp_decay"].append(temp_decay)
                            results["n_layers"].append(n_layers)
                            results["one_hot"].append(one_hot)
                        
                            result_file = open("grid_search_results/" + str(run_count) + ".out", "rb")
                            lines = result_file.readlines()
                            result_file.close()
                            
                            train_results = str(lines[-2]).split(' ')
                            val_results = str(lines[-1]).split(' ')
                            train_loss = float(train_results[4].split(',')[0])
                            val_loss = float(val_results[4].split(',')[0])
                            if loss == "mse": 
                                split_index = 6
                            else: 
                                split_index = 5
                            train_aux = float(train_results[split_index].split('\\')[0].split('=')[1].split(",")[0])
                            val_aux = float(val_results[split_index].split('\\')[0].split('=')[1].split(",")[0])
                            
                            results["loss_train"].append(train_loss)
                            results["loss_val"].append(val_loss)
                            results["aux_train"].append(train_aux)
                            results["aux_val"].append(val_aux)
                        
                            run_count +=1
```

```python
results = pd.DataFrame(results)
results_mse = results[results.loss_type == "mse"]
results_ce = results[results.loss_type == "cross-entropy"]
sorted_mse = results_mse.sort_values("loss_val")
sorted_ce = results_ce.sort_values("loss_val")
```
